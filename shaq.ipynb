```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
career = pd.read_csv('shaq.csv')
career2 = pd.read_csv('shaq2.csv')
playoffs = pd.read_csv('playshaq.csv')
```

```python
len(career)
```

```python
career2 = career2.drop('P/M', axis=1)
len(career2)
```

```python
shaq = pd.concat([career, career2])
len(shaq)
```

```python
shaq['type'] = np.ones(len(shaq)) * 0
shaq.head()
```

```python
playoffs = playoffs.drop('+/-', axis=1)
playoffs['type'] = np.ones(len(playoffs))
playoffs.head()
```

```python
def game_type(i):
    if i == 1:
        return 'Playoff'
    if i == 0:
        return 'Regular'
```

```python
shaq.head()
```

```python
shaq.columns
```

```python
playoffs.head()
```

```python
playoffs.columns
```

```python
columns = ['Tm', 'TRB', 'BLK', 'PTS', 'type']
```

```python
shaq = shaq[columns]
playoffs = playoffs[columns]
```

```python
cols = ['TRB', 'BLK', 'PTS']

for i in cols:
    shaq[i] = shaq[i].apply(int)
    playoffs[i] = playoffs[i].apply(int)
```

```python
s = shaq.groupby('Tm').mean()
```

```python
p = playoffs.groupby('Tm').mean()
```

```python
order = [6, 5, 2, 3, 1, 4]
```

```python
s['order'] = order
p['order'] = order
```

```python
p
```

```python
oneal = pd.concat([s, p])
oneal.head()
```

```python
oneal['Type'] = oneal['type'].apply(game_type)
oneal.head()
```

```python
sns.set_style("whitegrid", {'axes.grid' : False})
#sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
ax = sns.lineplot(x="order", y="PTS", hue="Type", data=oneal, marker='o')
#plt.setp(ax.get_legend().get_texts(), fontproperties=prop) # for legend text
#plt.setp(ax.get_legend().get_title(), fontproperties=prop)
prop = fm.FontProperties(fname='Liberator-Light.ttf')
plt.xticks(fontproperties=prop, size=14);
plt.yticks(fontproperties=prop, size=14);
plt.xlabel(' ', fontproperties=prop, size=12);
plt.ylabel('Points', fontproperties=prop, size=16);
ax.set_title('points by team', fontproperties=prop, size=20);
plt.setp(ax.get_legend().get_texts(), fontproperties=prop) # for legend text
plt.setp(ax.get_legend().get_title(), fontproperties=prop)

plt.tight_layout()
plt.savefig("shaqpt.png")
```
