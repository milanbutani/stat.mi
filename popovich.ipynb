```python
import pandas as pd
import numpy as np

from datascience import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nba18_19_team = pd.read_csv('teamdata.csv')
nba18_19_team.head()
```

```python
nba18_19_team[['Team', '3PA', '3P%']].plot.scatter('3P%', '3PA')
```

```python
sorted_3Ppercent = nba18_19_team.sort_values('3P%').tail(5)
sorted_3Ppercent
```

```python
#sorted_3Ppercent.plot.scatter('3P%', '3PA')

x = sorted_3Ppercent['3P%'].values
y = sorted_3Ppercent['3PA'].values
n = sorted_3Ppercent['Team'].values

fig, ax = plt.subplots()
ax.scatter(x, y)

for i, txt in enumerate(n, 0):
    ax.annotate(txt, (x[i], y[i]))
    
plt.savefig("top5.png")
```
