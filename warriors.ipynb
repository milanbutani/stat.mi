```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
teams = pd.read_csv('tms.csv')
teams.head()
```

```python
teams.columns
```

```python
df = teams[['Team', 'ORtg', 'DRtg', 'NRtg']]
df = df.append({'Team':'Bulls', 'ORtg':115.2, 'DRtg':101.8, 'NRtg':13.4}, ignore_index=True)
```

```python
sns.set(rc={'figure.figsize':(10,10)});
sns.set_style("whitegrid", {'axes.grid' : False});
ax = sns.scatterplot(x='ORtg', y='DRtg', data=df, color='darkseagreen');
prop = fm.FontProperties(fname='Liberator-Light.ttf');
plt.xticks(fontproperties=prop, size=14);
plt.yticks(fontproperties=prop, size=14);
plt.xlabel('Off. Rating', fontproperties=prop, size=12);
plt.ylabel('Def. Rating', fontproperties=prop, size=12);
plt.axhline(y=np.mean(df['DRtg']),linewidth=2, color='rebeccapurple');
plt.axvline(x=np.mean(df['ORtg']),linewidth=2, color='rebeccapurple');
ax.set_title('Offensive Vs. Defensive Rating', fontproperties=prop, size=20);
plt.savefig('warriors.png')
```
