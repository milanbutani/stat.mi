```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tat = pd.read_csv('nba-tattoos-data.csv')
tat.head()
```

```python
def first_letter(s):
    return s[0]

tat['letter'] = tat['Player Name'].apply(first_letter)
tat
```

```python
def boolean(s):
    if s == 'no':
        return 0
    if s == 'yes':
        return 1
    
tat['tat?'] = tat['Tattoos yes/no'].apply(boolean)
tat
```

```python
dataset = tat[['letter', 'tat?']]
grouped = dataset.groupby(by='letter').mean()
grouped = grouped.reset_index()
```

```python
grouped.columns = ['letter', 'with_tattoo']
grouped
```

```python
grouped['no_tattoo'] = 1 - grouped['with_tattoo']
grouped
```

```python
tat
```

```python
def last_name(s):
    names = s.split()
    return names[1]

last_name(tat['Player Name'][0])
```

```python
tat['Last Name'] = tat['Player Name'].apply(last_name)
tat['Last First Letter'] = tat['Last Name'].apply(first_letter)
tat
```

```python
last = tat[['tat?', 'Last First Letter']]
```

```python
last
```

```python
last_grouped = last.groupby(by='Last First Letter').mean()
last_grouped = last_grouped.reset_index()
last_grouped
```
