```python
insta = Table.read_table('insta.csv')
```

```python
insta= insta.relabel(' Instagram Follower Count (millions)', 'followers')
insta= insta.relabel(' Chips', 'chips')
insta= insta.relabel(' 2018-19 All Stars', '#allstars')
insta= insta.relabel(' All-NBA players', '#allnba')
insta= insta.relabel(' 2018-19 wins', '#wins')
insta
```

```python
import seaborn as sns

sns.scatterplot(insta['#wins'], insta['followers']);
plt.xlabel('#wins')
plt.ylabel('#followers(m)')
plt.title('Wins vs Followers (millions)')

plt.savefig("winvfollowers.png")
```
