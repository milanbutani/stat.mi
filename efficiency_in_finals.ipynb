```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
```

```python
finals = pd.read_csv('finals.csv')
finals.head()
```

```python
finals[finals['name'] == 'Isiah Thomas']
```

```python
def player(i):
    return i.split('\\')[0]
```

```python
finals['name'] = finals['Player'].apply(player)
finals.head()
```

```python
df = finals[['FG%', 'MP', 'name', 'FT%', '3P%']]
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style("whitegrid", {'axes.grid' : False})
ax = sns.scatterplot(x='MP', y='FG%', data=df, color='goldenrod');
prop = fm.FontProperties(fname='Liberator-Light.ttf')
plt.xticks(fontproperties=prop, size=14);
plt.yticks(fontproperties=prop, size=14);
plt.xlabel('Minutes Played', fontproperties=prop, size=12);
plt.ylabel('Field goal percentage', fontproperties=prop, size=12);
#plt.axhline(y=np.mean(data['Points']),linewidth=2, color='k');
ax.set_title('Minutes played vs FG% for finals mvp', fontproperties=prop, size=20)
#plt.plot([50, 450], [0, 150], '--', linewidth=2)
plt.tight_layout()
plt.savefig("fg.png")
```

```python
df[(df['FG%'] < 0.45) & (df['MP'] > 46)]
```
