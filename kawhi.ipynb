```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
```

```python
overview = pd.read_csv('kawhicareer.csv')
overview
```

```python
season = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']
two_distance = [6.28, 7.27, 7.87, 9.67, 9.76, 10.55, 9.64, 8.9, 9.81]
three_distance = [24.17, 23.79, 24.3, 24.52, 24.73, 25.05, 25.06, 25.33, 25.93]
```

```python
zipped = list(zip(season, two_distance, three_distance))
```

```python
distance = pd.DataFrame(columns=['season', '2dist', '3dist'], data=zipped)
```

```python
distance
```

```python
sns.set_style("whitegrid", {'axes.grid' : False})
ax = sns.lineplot(x='season', y='2dist', data=distance, color='red')
#plt.setp(ax.get_legend().get_texts(), fontproperties=prop) # for legend text
#plt.setp(ax.get_legend().get_title(), fontproperties=prop)
prop = fm.FontProperties(fname='Liberator-Light.ttf')
plt.xticks(fontproperties=prop, size=14);
plt.yticks(fontproperties=prop, size=14);
plt.xlabel('Season', fontproperties=prop, size=12);
plt.ylabel('Avg Shot Distance', fontproperties=prop, size=12);
#plt.axhline(y=np.mean(relevant['BLK']),linewidth=2, color=sns.color_palette("Set2")[0]);
#print('kawhi', np.mean(relevant['BLK']))
#plt.axhline(y=np.mean(wallace['BLK']),linewidth=2, color=sns.color_palette("Set2")[1]);
#print('wallace', np.mean(wallace['BLK']))
ax.set_title('Average 2-point Shot Distance Over time', fontproperties=prop, size=20)

plt.tight_layout()
plt.savefig("2shot.png")
```
